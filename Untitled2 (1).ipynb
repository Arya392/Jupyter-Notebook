```python
# Find only columns with missing values
missing_cols = df.columns[df.isna().any()].tolist()
```

```python
num_missing_cols = df[missing_cols].select_dtypes(include=[np.number]).columns.tolist()
```

```python
cat_missing_cols = df[missing_cols].select_dtypes(include=['object']).columns.tolist()
```

```python
print("Numeric with missing:", num_missing_cols)
print("Categorical with missing:", cat_missing_cols)
```

```python
# Fill missing values
for col in num_missing_cols:
    df.loc[:, col] = df[col].fillna(df[col].median())

```

```python
for col in cat_missing_cols:
    df.loc[:, col] = df[col].fillna("missing")
```

```python
# Check again
print(df.isna().sum().sort_values(ascending=False).head(10))
```

```python
df['target'] = df['Decision'].map({'APPROVED': 1, 'DECLINED': 0})
```

```python
print(df['target'].value_counts())
```

```python
# Features (X) and target (y)
X = df.drop(columns=['Decision', 'target'])
y = df['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=25)

```

```python
cat_cols = X.select_dtypes(include=['object', 'category']).columns.tolist()
num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
```

```python
low_card_cols = [c for c in cat_cols if X[c].nunique() <= 20]
high_card_cols = [c for c in cat_cols if X[c].nunique() > 20]
```

```python
preprocess = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('low_cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), low_card_cols),
        ('high_cat', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1), high_card_cols),
    ],
    remainder='drop'
)
```

```python
minority_count = y_train.value_counts().min()
k = int(max(1, min(5, minority_count - 1)))
```

```python

pipe = ImbPipeline(steps=[
    ('prep', preprocess),
    ('smote', SMOTE(random_state=42, k_neighbors=k)),
    ('clf', LGBMClassifier(
        objective='binary',
        learning_rate=0.05,
        n_estimators=800,
        num_leaves=64,
        subsample=0.8,
        colsample_bytree=0.8,
        reg_lambda=1.0,
        n_jobs=-1,
        random_state=42
    ))
])
```

```python
pipe.fit(X_train, y_train)
```

```python
y_proba = pipe.predict_proba(X_test)[:, 1]
```

```python
def best_threshold_macro_recall(y_true, proba):
    best_t, best_score = 0.5, -1.0
    for t in np.linspace(0.05, 0.95, 181):
        y_hat = (proba >= t).astype(int)
        r0 = recall_score(y_true, y_hat, pos_label=0)
        r1 = recall_score(y_true, y_hat, pos_label=1)
        score = 0.5 * (r0 + r1)
        if score > best_score:
            best_score, best_t = score, t
    return best_t, best_score
```

```python
best_t, macro_recall = best_threshold_macro_recall(y_test, y_proba)
print(f"Best threshold for macro recall: {best_t:.3f} (macro recall={macro_recall:.3f})")
```

```python
y_pred_opt = (y_proba >= best_t).astype(int)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_opt))
print("AUC:", roc_auc_score(y_test, y_proba))
print(classification_report(y_test, y_pred_opt, digits=3))
print("Confusion matrix @ best threshold:")
print(confusion_matrix(y_test, y_pred_opt))
```
